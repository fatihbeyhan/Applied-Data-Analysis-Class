```python
import pandas as pd
from sklearn.model_selection  import cross_val_score, GridSearchCV, train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, \
                            precision_score,recall_score, auc, roc_auc_score, roc_curve
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
import xgboost as xgb
import seaborn as sns
import statsmodels.api as sm
import numpy
import numpy as np
import itertools
import warnings
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
warnings.filterwarnings("ignore")
```

```python
dataset_name = 'ICUPatients.csv'
data = pd.read_csv(dataset_name)
categorical_columns = list(data.columns[data.dtypes == 'object'])
data = pd.get_dummies(data) # One Hot Encoding
train_y_rd = data["Readmitted"]
train_y_dt = data["death"]
train_x = data.drop('death', axis=1)
train_x = train_x.drop('Readmitted',axis=1)
train_x = train_x.drop('Patientid',axis=1)
train_x = pd.DataFrame(train_x,columns=train_x.columns)
train_x_columns = train_x.columns
```

```python
def forward_selection(data, target, significance_level=0.05):
    initial_features = data.columns.tolist()
    best_features = []
    while (len(list(initial_features))>0):
        remaining_features = list(set(initial_features)-set(best_features))
        new_pval = pd.Series(index=remaining_features)
        for new_column in remaining_features:
            model = sm.OLS(target, sm.add_constant(data[best_features+[new_column]])).fit()
            new_pval[new_column] = model.pvalues[new_column]
        min_p_value = new_pval.min()
        if(min_p_value<significance_level):
            best_features.append(new_pval.idxmin())
        else:             
            break
    return best_features
```

```python
best_features_dt = forward_selection(train_x,train_y_dt)
```

```python
best_features_rd = forward_selection(train_x,train_y_rd)
```

```python
train_dt = train_x[best_features_dt]
```

```python
train_rd = train_x[best_features_rd]
```

```python
train_split_x_rd, test_split_x_rd, train_split_y_rd, test_split_y_rd = train_test_split(train_rd, train_y_rd, test_size=0.2, random_state=0)
train_split_x_dt, test_split_x_dt, train_split_y_dt, test_split_y_dt = train_test_split(train_dt, train_y_dt, test_size=0.2, random_state=0)
```

## KNN For Readmitted Based Data

```python
_range_ = [i for i in range(1,30)]
```

```python
knn_training_accuracy_scores_rd = []
knn_test_accuracy_scores_rd = []
for k in range(1,30):
        knn = KNeighborsClassifier(n_neighbors = k)
        knn.fit(train_split_x_rd,train_split_y_rd)
        
        preds_for_training = knn.predict(train_split_x_rd)
        preds_for_test = knn.predict(test_split_x_rd)
        
        accuracy_scores_for_training = accuracy_score(preds_for_training,train_split_y_rd)
        knn_training_accuracy_scores_rd.append(accuracy_scores_for_training)
        
        accuracy_scores_for_test = accuracy_score(preds_for_test,test_split_y_rd)
        knn_test_accuracy_scores_rd.append(accuracy_scores_for_test)
```

```python
plt.plot(_range_,knn_training_accuracy_scores_rd,color='green')
plt.plot(_range_,knn_test_accuracy_scores_rd,color='red')
plt.xlabel('K Neighbors For KNN', fontsize=18)
plt.ylabel('Accuracy', fontsize=18)
plt.title("Readmitted Based Data KNN Accuracy")
plt.legend(labels=["KNN Training Accuracy","KNN Test Accuracy"],bbox_to_anchor=(0., 1.09, 1., 0.), handletextpad=0.3,
                    loc='lower left', ncol=6, mode="expand", borderaxespad=0,
                    numpoints=1, handlelength=0.5)
```
